### 히스토그램 hist() 함수 사용

```python
import csv
import matplotlib.pyplot as plt

plt.figure(dpi=200)  # 해상도

# 히스토그램 -> 도수분포 -> 어떠한 자료에 속해 있는 값들을 수량으로 나타낸 것
# 보통 연령대들을 많이 표시한다.
# 데이터의 불균형도를 보는것, 평균적으로 데이터가 잘 구축되어 있는지 확인
# 플롯 차트보다 히스토그램을 돌려서 먼저 확인하는 경우가 많다.
plt.hist([1,2,3,4,5,6,6,7,8,10])  # hist(dataList) -> 가로축은 계급(아이템), 세로축은 도수(횟수, 빈도수) 
plt.show()
```

### 1월과 8월의 데이터를 히스토그램으로 시각화 하기

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 초고 기온을 담을 jan 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출

  if row[-1] != '':
    if month == '08' :  # 8월 데이터이면 aug 변수에 담기
      aug.append(float(row[-1]))

    if month == '01' : # 1월 데이터이면 jan 변수에 담기
      jan.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(aug, bins=100, color='r', label='Aug') 
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend() # 두개 이상 비교할 때는 항상 label을 추가하고, plt.legend()를 선언 해줘야 한다.
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 초고 기온을 담을 jan 리스트 변수 선언
april = [] # 6월의 최고 기온을 담을 

for row in data:
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출

  if row[-1] != '':
    if month == '08' :  # 8월 데이터이면 aug 변수에 담기
      aug.append(float(row[-1]))

    if month == '01' : # 1월 데이터이면 jan 변수에 담기
      jan.append(float(row[-1]))

    if month == '04' : 
      april.append(float(row[-1]))

# 최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(aug, bins=100, color='r', label='Aug') 
plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(april, bins=100, color='g', label='Apr')
plt.legend() # 두개 이상 비교할 때는 항상 label을 추가하고, plt.legend()를 선언 해줘야 한다.
plt.show()
```

### 1월과 8월의 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

jan = [] # 최고 기온 데이터를 지정할 리스트 생성
aug = []

for row in data :
  if row[-1] != '' :
    if int(row[0].split('-')[1]) == 1 :
      jan.append(float(row[-1]))

    if int(row[0].split('-')[1]) == 8 :
      aug.append(float(row[-1]))
    

# 박스플롯 시각화
plt.figure(dpi=200)
plt.boxplot([jan, aug])
plt.show()
```

### 1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기


# month= [[],[],[],[],[],[],[],[],[],[],[],[]] # 월별 데이터 저장할 month 변수 선언
month = []
for i in range(12):
  month.append([])

for row in data:
  if row[-1] != '' :
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 리스트 안에 리스트의 방번호를 지정해서 리스트에 담기
    # 월과 같은 번호의 인덱스에 월별 데이터 저장(예: 1월 row[0].split('-')[1]-1 -> month[0])

# 박스플롯 시각화
plt.figure(dpi=200)
plt.boxplot(month)
plt.show()


```

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])  # day리스트 내에 31개 리스트 생성

for row in data :
  if row[-1] != '' :
    if row[0].split('-')[1] == '08':  
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))  # 8월 최고 기온 값 저장 day[0] -> 1일

# 박스플롯 시각화
plt.figure(dpi=200)
plt.boxplot(day)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])  # day리스트 내에 31개 리스트 생성

for row in data :
  if row[-1] != '' :
    if row[0].split('-')[1] == '08':  # 8월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

# 박스플롯 시각화
plt.style.use('ggplot')  # 그래프 스타일 지정(회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,2), dpi=200)
plt.boxplot(day, showfliers=False) # 아웃라이어(이상치) 생략하기
plt.show()
```
