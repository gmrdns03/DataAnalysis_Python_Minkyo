### 데이터 준비하기

```python
# 최고 기온 데이터만 리스트에 담아서 출력하기

import csv

f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

result = [] # 최고 기온 데이터를 지정할 리스트 생성

for row in data :
  if row[-1] != '' :
    result.append(float(row[-1]))

print(result)  # 최고 기온 리스트
```

```python
# 수능날 기온 데이터 시각화
import csv
import matplotlib.pyplot as plt

f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

suneng = ['16','23','22','13','01','18','17','15','07','06','05','17','23','16','15','13','02','18','10','08','07','13','12','17','16','23']
high = [] # 최고 기온
low = [] # 최저 기온

# 둘 간의 차이를 비교하는 것이기 때문에 둘 중 어느 하나가 존재하더라도 결측치가 되어버린다.
for i in range(len(suneng)):
  for row in data:
    if row[-1] != '' and row[-2] != '' :
      if 1994 <= int(row[0].split('-')[0]) :
        if row[0].split('-')[1] == '11' and row[0].split('-')[2] == suneng[i] :
          high.append(float(row[-1])) # 최고 기온 리스트
          low.append(float(row[-2])) # 최저 기온 리스트

# 데이터 시각화
plt.figure(figsize = (15,2), dpi = 200)  # 해상도 200
plt.plot(list(range(1994, 2019)),high, color='hotpink', marker='*', label='high')  # 최고기온
plt.plot(list(range(1994, 2019)),low, color='skyblue', marker='*', label='low')  # 최저기온
plt.xticks(list(range(1994, 2019)))
plt.legend()  # 범례표시
plt.show()
```

```python
# 겨울 기온 시각화 문제 해결
import csv
import matplotlib.pyplot as plt

f =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

high = [] # 최고 기온
avg = [] # 평균 기온
low = [] # 최저 기온

# 둘 간의 차이를 비교하는 것이기 때문에 둘 중 어느 하나가 존재하더라도 결측치가 되어버린다.
for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '' :  # 최고, 최저, 평균 결측치 제거
    if 1997 <= int(row[0].split('-')[0]) :
      if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24' :
        high.append(float(row[-1])) # 최고 기온 리스트
        avg.append(float(row[-3]))
        low.append(float(row[-2])) # 최저 기온 리스트

# 데이터 시각화
plt.figure(dpi = 200)  # 해상도 200
plt.plot(high, 'hotpink', label='high')  # 최고기온
plt.plot(low, 'skyblue', label='low')  # 최저기온
plt.plot(avg, 'g', label='avg')  # 최저기온

plt.title('크리스마스 이브 날씨 변화')
plt.legend()  # 범례표시

plt.show()
```
