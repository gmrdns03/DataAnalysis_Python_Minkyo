### plt.plot(데이터셋) 그래프 그리기

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기
```

```python
plt.plot([10, 20, 30, 40])  # y값이 디폴트/ y값이 주어지면, 자동으로 x값을 0부터 1씩 증가함
# 머신러닝 하다보면 y가 가장 중요하다.
plt.show()
```

### plt.plot(x, y) 그래프 그리기

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.plot([1,2,3,4], [12, 43, 25, 15])  # plt.plot(x, y) 함수에 데이터 입력하기

plt.show()  # 그래프 보여주기
```

### plt.plot(x, y)에 타이틀 달아주기

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('plotting') # 타이틀 속성 추가 (한글은 깨짐)
plt.plot([1,2,3,4], [12, 43, 25, 15])  # plt.plot(x, y) 함수에 데이터 입력하기

plt.show() 
```

### 두개의 데이터를 넣어서 그래프 그리기

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('plotting') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], label='asc') # 증가를 의미하는 asc
plt.plot([40,30,20,10], label='desc')  # 감소를 의미하는 desc

plt.legend(loc=9)  # 그래프에 범례 속성 추가
# legend의 위치 (loc=) / 디폴트는 5
# 2  9  1
# 6 10  5,7
# 3  8  4
plt.show() 
```

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('plotting') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], color='skyblue', label='skyblue') # 하늘색 그래프
plt.plot([40,30,20,10], 'pink', label='pink')  # 분홍색 그래프

plt.legend(loc=5)  # 그래프에 범례 속성 추가

plt.show() 
```

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('plotting') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], color='r', label='red') # 빨간색 그래프
plt.plot([40,30,20,10], color='k', label='black')  # 분홍색 그래프

# 항상 명확하게 
# 그래프의 색깔 color=''에 약자 넣기/ 기본 색상은 약자로 표시 가능
# r= red/ g= green/ b= blue/ k= black/ y= yellow

plt.legend(loc=5)  # 그래프에 범례 속성 추가

plt.show() 
```

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('linestyle') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], color='r', linestyle='--', label='dashed') # 빨간색 dashed 그래프
plt.plot([40,30,20,10], color='k', ls=':', label='black')  # 검정색 점선 그래프
# linestyle을 ls라고 약칭해도 된다.

plt.legend(loc=5)  # 그래프에 범례 속성 추가

plt.show() 
```

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('marker') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], 'r.', ls='--', label='circle') # 빨간색 dashed 그래프
plt.plot([40,30,20,10], 'k^', ls=':', label='triangle up')  # 검정색 점선 그래프

plt.legend(loc=5)  # 그래프에 범례 속성 추가

plt.show() 
```

```python
import matplotlib.pyplot as plt  # 라이브러리 불러오기

plt.title('marker') # 타이틀 속성 추가 (한글은 깨짐)

# 레이블 속성은 범례를 달아줄 수 있다.
plt.plot([10,20,30,40], 'r.-', label='cicle line') # 빨간색 dashed 그래프
plt.plot([40,30,20,10], 'k^--', label='triangle dashed')  # 검정색 실선 그래프
# 색상과 선 모양, 마커 모양을 동시에 설정하고 싶을 때는
# '<색상><마커모양><선모양>'순으로 코드를 작성하면 된다.

plt.legend(loc=5)  # 그래프에 범례 속성 추가

plt.show() 
```

```python
import csv
import matplotlib.pyplot as plt  # 1) 라이브러리 불러오기

f = open('/content/seoul_temp_19040101_20190117.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)  # 첫 번째 행을 읽은 후 커서를 다음 행으로 이동시키는 함수

result = []

for row in data :
  if row[-1] != '' :  # 최고 기온이 ''(빈문자열)이 아니면, result리스트에 넣는다.
    result.append(float(row[-1]))

result

plt.figure(figsize=(10,2), dpi=200)  # dpi=해상도
plt.plot(result, 'r')
plt.show()

# -값들이 사라졌다.
```
