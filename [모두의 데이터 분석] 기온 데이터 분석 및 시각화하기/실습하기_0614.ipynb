### 2) 상자그림을 활용한 최고 기온 vs 서울 최고기온

```python
import csv
import matplotlib.pyplot as plt


f1 =open('/content/deakwanryung_temp_.csv', encoding='cp949')
f2 =open('/content/seoul_temp_19040101_20190117.csv', encoding='cp949')
data1 = csv.reader(f1) # 대관령 데이터 읽기
data2 = csv.reader(f2) # 서울 데이터 읽기
next(data1) # 헤더 넘기기
next(data2)

high_seoul = []
high_de = []

for row in data1 :
  if row[-1] != '' :
    high_seoul.append(float(row[-1]))

for row in data2 :
  if row[-1] != '' :
    high_de.append(float(row[-1]))

# 박스플롯 시각화
plt.style.use('ggplot')
plt.figure(figsize=(2,2), dpi=200)
plt.boxplot([high_seoul, high_de])
plt.show()

f1.close()
f2.close()
```

### 3) 월별 최고기온 상자 그림 그리기

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/deakwanryung_temp_.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

month = []
for i in range(12):
  month.append([])

for row in data :
  if row[-1] != '' :
    month[int(row[0].split('-')[1])-1].append(float(row[-1]))

# 박스플롯 시각화
plt.style.use('ggplot')
plt.figure(figsize=(5,2), dpi=200)
plt.boxplot(month, showfliers=False)
plt.show()
```

### 4) 특정 달의 일별 최고 기온 상자그림 그리기

```python
import csv
import matplotlib.pyplot as plt


f =open('/content/deakwanryung_temp_.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 넘기기

day = [] # 일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31):
  day.append([])  # day리스트 내에 31개 리스트 생성

for row in data :
  if row[-1] != '' :
    if row[0].split('-')[1] == '01':  # 1월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))

# 박스플롯 시각화
plt.style.use('ggplot')
plt.figure(figsize=(10,2), dpi=200)
plt.boxplot(day, showfliers=False)
plt.show()
```
