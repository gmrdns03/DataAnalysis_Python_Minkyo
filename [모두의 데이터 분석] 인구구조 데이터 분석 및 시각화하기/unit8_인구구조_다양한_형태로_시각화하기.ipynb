## 지역명을 입력받아서 연령별 인구수 데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

result = []

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :  ')

for row in data :
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

# 인구구조 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

plt.title(name+'지역의 인구 구조')
plt.plot(result)
plt.show()

f.close()
```

### 수평막대그래프(barh()함수)

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

result = []

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :  ')

for row in data :
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

# 인구구조 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

plt.title(name+'지역의 인구 구조')
plt.barh(range(101),result)
plt.show()

f.close()

# 연령을 강조하고 싶으면 수직 막대그래프
# 명수를 강조하고 싶으면 수평 막대그래프로, 양 극단의 값을 비교하고자 할때
# 어떤 것을 강조하고 싶은지에 따라서 막대그래프의 모양을 선택해야 한다.
# 보통 사람들은 X축을 먼저 본다.
```

#### 성별 데이터 저장하기 (Type 2)

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

m = []
fe = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(int(i))
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j))

print(m)
print(fe)

f.close()
    

# 코딩에 왕도는 없지만 간결하고 보기 좋은 방식으로 코딩하는 것이 좋다.
# 가독성이 좋은 방식이 좋음
```

#### 범례 추가하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

m = []
fe = []

for row in data:
  if '신도림' in row[0]:
    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(-(int(i)))  # 항아리 차트를 그리기 위해 음수로 변환 시키는 작업을 먼저 해준다.
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j))


plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101),m, label='남성')
plt.barh(range(101),fe, label='여성')
plt.legend()
plt.show()
```
