```python
# 1. 수원의 성별 데이터(항아리 그래프)
import csv
import matplotlib.pyplot as plt


f = open('/content/gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

m = []
f = []

for row in data:
  if '율천동' in row[0]:
    for i in range(0, 101):
      m.append(-int(row[i + 3])) # 값들이 음수로 들어간다.
      f.append(int(row[-(i+1)])) # -1 값부터 담는다. -> 100세 이상부터 값이 담긴다.



f.reverse() 
# 0세부터 인구수 나올 수 있도록 데이터 뒤집기

print(m)
print(f)

plt.figure(dpi=200)
plt.style.use('ggplot')  # 격자무늬 스타일로 지정  classic, dark_background, ggplot
plt.rcParams['axes.unicode_minus'] = False # x축의 마이너스 ticks를 알려준다. 
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.title('율천동 지역의 남녀 성별 인구 분포')
plt.show()

# 수원시 장안구 율천동에는 성균관대 캠퍼스가 있음
# 20대 초반 남성의 비율이 여성의 비율보다 확연하게 크다는 사실을 알 수 있음
# 수원 성균관대는 이과캠퍼스이고 체대가 있어서 남성의 비율이 크다는 것을 추측할 수 있음 
```

```python
# 2. 2019년 3월 연령별 인구 구조 -> 2020년 3월 연령별 인구 구조 비교(같이 그리기) 

import csv
import matplotlib.pyplot as plt




f_2019 = open('/content/age_2019_03.csv', 'r', encoding='cp949')
data_2019 = csv.reader(f_2019)
header_2019 = next(data_2019)

result2019 = []

for row_2019 in data_2019:
  if '율천동' in row_2019[0]:
    for i_2019 in row_2019[3:]:
      new_2019 = i_2019.replace(',', '')
      result2019.append(int(new_2019))  # 인구수 데이터 정수로 변환한 후 result 변수에 담기

f_2020 = open('/content/age_2020_03.csv', 'r', encoding='cp949')
data_2020 = csv.reader(f_2020)
header_2020 = next(data_2020)

result2020 = []

for row_2020 in data_2020:
  if '율천동' in row_2020[0]:
    for i_2020 in row_2020[3:]:
      new_2020 = i_2020.replace(',', '')
      result2020.append(int(new_2020))  # 인구수 데이터 정수로 변환한 후 result 변수에 담기


plt.figure(dpi=200)
plt.style.use('ggplot')  # 격자무늬 스타일로 지정  classic, dark_background, ggplot
plt.plot(result2019, label='2019년 율천동 연령별 인구분포')
plt.plot(result2020, label='2020년 율천동 연령별 인구분포')
plt.title('율천동 지역의 연령별 인구 분포')
plt.legend()
plt.show()

# 코로나로 인해 20대의 비율이 줄어들 것이라고 예상을 세웠음
# 하지만 오히려 늘어났음
# 수원 성대는 이과, 체대이므로 원격으로 수업이 불가능한 과들이 많이 있어서 코로나의 영향을 덜 받은 것으로 예상함
```
