## 제주특별 자치도의 성별 분포를 항아리 모양 그래프로 표현

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('지역명을 입력해주세요 :')

m = []
fe = []

for row in data:
  if name in row[0]:
    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(-(int(i.replace(',',''))))  # 항아리 차트를 그리기 위해 음수로 변환 시키는 작업을 먼저 해준다.
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j.replace(',','')))

# 시각화 설정
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title(name+'지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101),m, label='남성')
plt.barh(range(101),fe, label='여성')
plt.legend()
plt.show()
```

## 제주도의 성별 인구 파이차트로 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('찾고 싶은 지역명을 입력해주세요 :')

size = [] # 제주도의 남성 인구 합계

m = 0  # 남성 인구수 합계 변수
fe = 0  # 여성 인구수 합계 변수

for row in data:
  if name in row[0]:
   
    for i in range(101) :
      m += int(row[3+i].replace(',', ''))  # m = m + int(row(3+i).replace(',', ''))
      fe += int(row[106+i].replace(',', ''))  # fe = fe + int(row(3+i).replace(',', ''))
    break

    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(-(int(i.replace(',',''))))  # 항아리 차트를 그리기 위해 음수로 변환 시키는 작업을 먼저 해준다.
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j.replace(',','')))
    break  # 첫번째 데이터만 확인하고 탈출하기 위해서

size.append(m)
size.append(fe)


# 파이차트 시각화
plt.style.use('ggplot')
plt.figure(dpi=200)
color = ['crimson', 'darkcyan']
label = ['남', '여']
plt.title(name+'지역의 남녀 성별 비율')

plt.axis('equal')  # <- 범례를 옆으로 빼는 함수
plt.pie(size, labels=label, autopct='%.1f%%', explode=(0,0.1), colors=color, startangle=90) # 혈액형의 비율을 임의로 작성 / 'autopct=' 퍼센트 보이기 
plt.legend()
plt.show()
```
