# 1) 성별 데이터 수평 막대 그래프로 시각화 (항아리 그래프)
-> 본인 동네에 대해서

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

m = []
fe = []

for row in data:
  if '일산서구 탄현동' in row[0]:
    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(-(int(i)))  # 항아리 차트를 그리기 위해 음수로 변환 시키는 작업을 먼저 해준다.
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j))

# 시각화 설정
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('일산서구 탄현동 지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
plt.barh(range(101),m, label='남성')
plt.barh(range(101),fe, label='여성')
plt.legend()
plt.show()
```

# 2) 2019년 2월 연령별 인구구조/2020년 2월 인구구조 비교 (같이 그리기) _플롯차트

```python
import csv
import matplotlib.pyplot as plt

f19 = open('/content/age2019_02.csv', 'r', encoding='cp949')
f20 = open('/content/age2020_02.csv', 'r', encoding='cp949')
data_19 = csv.reader(f19)
data_20 = csv.reader(f20)
next(data_19)
next(data_20)

result_19 = []
result_20 = []

for row in data_19 :
  if '일산서구 탄현동' in row[0]:
    for i in row[3:]:
      result_19.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

for row in data_20 :
  if '일산서구 탄현동' in row[0]:
    for i in row[3:]:
      result_20.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

# 시각화하기
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')
plt.title('일산서구 탄현동 지역의 연령별 인구 분포')
plt.plot(result_19, label='2019년 일산서구 탄현동 연령별 인구 분포')
plt.plot(result_20, label='2020년 일산서구 탄현동 연령별 인구 분포')
plt.legend()
plt.show()
```
