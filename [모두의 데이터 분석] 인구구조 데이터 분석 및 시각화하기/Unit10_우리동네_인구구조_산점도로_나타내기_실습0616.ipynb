## 8개의 시도의 남성 여성의 인구 차이 그래프 그려서 비교하기<br>
부산광역시<br>
대구광역시<br>
인천광역시<br>
광주광역시<br>
대전광역시<br>
울산광역시<br>
서울특별시<br>
세종특별자치도


```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('지역명을 입력해주세요 :')

result = []  # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

for row in data:
  if name in row[0] :
    for i in range(3, 104):
      result.append(int(row[i].replace(',',''))- int(row[i+103].replace(',','')))
    break

# 시각화 설정
plt.style.use('ggplot')
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 꺽은선 그래프로 그리기
plt.title(name+' 지역의 남녀 성별 인구 차이')
plt.bar(range(101), result)
# plt.legend()
plt.show
```

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/age_gender2019_02.csv", encoding="cp949")
data = csv.reader(f)
result = []
location = ["서울특별시","부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종"]

#name=input("찾고 싶은 지역의 이름을 알려주세요 :")
for name in location:
    globals()[f"result_{name}"] = []
    for row in data: 
        if name in row[0]:
            for i in range(3, 104):
                globals()[f"result_{name}"].append(int(row[i].replace(",","")) - int(row[i+103].replace(",","")))
            break

fig = plt.figure(dpi=100, figsize=(10,16)) 
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')

for idx, name in enumerate(location):
    globals()[f"plot_{name}"] = fig.add_subplot(4, 2 ,idx+1)
    globals()[f"plot_{name}"].bar(range(101),globals()[f"result_{name}"])
    globals()[f"plot_{name}"].set_title(f"{name}")
plt.legend()
plt.show()

```

```python
import csv
import matplotlib.pyplot as plt

f = open("/content/age_gender2019_02.csv", encoding="cp949")
data = csv.reader(f)
result = []
location = ["서울특별시","부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종특별자치시"]

#name=input("찾고 싶은 지역의 이름을 알려주세요 :")
for name in location:
    globals()[f"result_{name}"] = []
    for row in data: 
        if name in row[0]:
            for i in range(3, 104):
                globals()[f"result_{name}"].append(int(row[i].replace(",","")) - int(row[i+103].replace(",","")))
            break

fig = plt.figure(dpi=75, figsize=(20,14)) 
plt.rc('font', family = 'NanumBarunGothic')
plt.style.use('ggplot')

for idx, name in enumerate(location):
    globals()[f"plot_{name}"] = fig.add_subplot(2, 4 ,idx+1)
    globals()[f"plot_{name}"].bar(range(101),globals()[f"result_{name}"])
    globals()[f"plot_{name}"].set_ylim([-10000,4000])
    globals()[f"plot_{name}"].set_title(f"{name}")
plt.legend()
plt.show()
```
