## 꺽은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('지역명을 입력해주세요 :')

m = []
fe = []

for row in data:
  if name in row[0]:
    for i in row[3:104] :  # 남성 데이터 컬럼 3~103 개의 데이터를 슬라이싱 - 정수 데이터로 변환 -> m리스트에 저장
      m.append(int(i.replace(',','')))  # 항아리 차트를 그리기 위해 음수로 변환 시키는 작업을 먼저 해준다.
    for j in row[106:] :  # 여성 데이터 : 106~ 끝까지 데이터 슬라이싱 -> 정수 데이터로 변환 -> fe리스트에 저장
      fe.append(int(j.replace(',','')))
    break

# 시각화 설정
plt.style.use('ggplot')
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')

# 꺽은선 그래프로 그리기
plt.plot(m, label='남성')
plt.plot(fe, label='여성')
plt.legend()
plt.show
```

## 연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('지역명을 입력해주세요 :')

result = []  # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

for row in data:
  if name in row[0] :
    for i in range(3, 104):
      result.append(int(row[i].replace(',',''))- int(row[i+103].replace(',','')))
    break

# 시각화 설정
plt.style.use('ggplot')
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 꺽은선 그래프로 그리기
plt.title(name+'지역의 남녀 성별 인구 차이')
plt.bar(range(101), result)
plt.legend()
plt.show
```

### 원의 크기는 인구수의 제곱근 추가

데이터가 너무 커서 차트로 그릴 때 비교가 어렵다면 루트를 씌워서 데이터를 표현하면 보기 좋아질 수 있다.

```python
############### 연령대별 성별 비율 산점도로 표현하기 #######################

import csv
import matplotlib.pyplot as plt
import math  # 제곱근을 구하기 위해 math라는 라이브러리 추가

f = open('/content/age_gender2019_02.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('지역명을 입력해주세요 :')

m = []
fe = []
size = [] # 인구수 사이즈 리스트 변수 선언

for row in data:
  if name in row[0]:
    for i in row[3:104] :  
      m.append(int(i.replace(',',''))) 
      # 제곱근 함수 math.sqrt()를 사용하여 거대한 인구수 데이터를 제곱근을 통해 작게 만들어 준다.  -> 스케일링 작업
      size.append(math.sqrt(int(i.replace(',',''))))  # 해당 연령대의 남성 인구를 size라는 리스트에 넣는다.
    for j in row[106:] : 
      fe.append(int(j.replace(',','')))
      # size.append(math.sqrt(int(j.replace(',',''))))
    break
# 인구수와 math.sqrt(인구수)차이
print(m)
print(size)
# 시각화 설정
plt.style.use('ggplot')
plt.figure(dpi = 200)
plt.rc('font', family = 'NanumBarunGothic')

# 산점도 그래프로 남여 인구수 데이터 표현하기
plt.title(name+' 지역의 산점도와 추세선')
plt.scatter(m, fe, s=size, c=range(101), cmap='jet', alpha=0.5)  # s=size를 선언하여 남성 인구의 크기 별로 
plt.plot(range(max(m)), range(max(m)), 'g')  # 
plt.xlabel('남성')
plt.ylabel('여성')
plt.colorbar()
plt.show()

# 연령의 정보를 알 수가 없다.
```
