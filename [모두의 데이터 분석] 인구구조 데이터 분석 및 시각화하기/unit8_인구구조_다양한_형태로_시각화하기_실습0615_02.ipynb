```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

result = []

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :  ')

for row in data :
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

# 인구구조 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

plt.title(name+'지역의 인구 구조')
plt.plot(result)
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/age.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

result = []

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 :  ')

for row in data :
  if name in row[0]:
    for i in row[3:]:
      result.append(int(i.replace(",", ""))) # 인구수 데이터 정수로 변환 후 result 변수에 담기

# 인구구조 시각화
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')

plt.title(name+'지역의 인구 구조')
plt.barh(range(101),result)
plt.show()

f.close()

# shape mismatch: object cannot be boradcast to a single shape 오류가 났을 경우 축에 데이터 가 서로 맞아떨어지지 않는 것
# 축에 들어가는 정보를 확인해야 한다.
```
