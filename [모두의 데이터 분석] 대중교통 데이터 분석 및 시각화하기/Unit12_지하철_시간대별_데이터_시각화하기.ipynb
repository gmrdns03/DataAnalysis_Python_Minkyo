## 헤더 제외하고 map함수를 활용한 데이터 타입 바꾸기(문자열 -> 정수)

map은 원본 리스트를 변경하지 않고 새 리스트를 생성합니다

list(map(int, 리스트))

list(map(str, range(10)))

 a = map(int, input().split()) <br>
10 20 (입력) <br>
 a <br>
 
<map object at 0x03DFB0D0> <br>
 list(a) <br>
[10, 20]

## 밤 11시에 사람들이 가장 많이 타는 역은 어디일까


승차 패턴 찾기<br> 
승차 시각이 1시간씩 늦어질 때 마다 인덱스 2씩 늘어남 <br> 승차 시작 시간의 인덱스 값 4 -> 4시부터 승차가 시작(첫차)<br> 
승차 시각(t) 와 승차 인원이 적힌 인덱스(i) 관계 분석 <br>
i = 4 + (t -4) * 2


```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwayTime.csv', 'r') # 인코딩이 utf-8이라면 생략 가능
data = csv.reader(f)
header_01 = next(data)
header_02 = next(data)

mx = 0
mx_station = ''

t = int(input('승차의 시간을 입력해주세요:  '))

for row in data : 
  row[4:] = map(int, row[4:])
  a = row[4 + (t-4)*2]  # 입력받은 시각의 승차 인원 값 추출

  if a > mx :
    mx = a
    mx_station = row[3] + '(' + row[1] + ')'

print(mx_station, mx)
```

## 시간대별 최대 승차 역 이름 및 승차 인원 막대 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwayTime.csv', 'r') # 인코딩이 utf-8이라면 생략 가능
data = csv.reader(f)
header_01 = next(data)
header_02 = next(data)

mx = [0] *24
mx_station = [''] * 24

for row in data : 
  row[4:] = map(int, row[4:])
  for j in range(24) :
    a = row[j * 2 + 4]  # 시간대별 패턴 i = j * 2 + 4 (j는 0~23까지) -> 시작 인덱스 4
    if a > mx[j] :
      mx[j] = a 
      mx_station[j] = row[3]

plt.rc('font', family = 'NanumBarunGothic')
plt.figure(dpi = 200)
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwayTime.csv', 'r') # 인코딩이 utf-8이라면 생략 가능
data = csv.reader(f)
header_01 = next(data)
header_02 = next(data)

mx = [0] *24
mx_station = [''] * 24

for row in data : 
  row[4:] = map(int, row[4:])
  for j in range(24) :
    a = row[j * 2 + 4]  # 시간대별 패턴 i = j * 2 + 4 (j는 0~23까지) -> 시작 인덱스 4
    if a > mx[j] :
      mx[j] = a 
      mx_station[j] = row[3] + '(' + str(j+4) + ')' # 역 이름과 시간대
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.figure(dpi = 200)
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

## 시간대별로 하차 인원이 가장 많은 역 막대그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwayTime.csv', 'r') # 인코딩이 utf-8이라면 생략 가능
data = csv.reader(f)
header_01 = next(data)
header_02 = next(data)

mx = [0] *24
mx_station = [''] * 24

for row in data : 
  row[4:] = map(int, row[4:])
  for j in range(24) :
    a = row[j * 2 + 5]  # 시간대별 패턴 i = j * 2 + 4 (j는 0~23까지) -> 시작 인덱스 4
    if a > mx[j] :
      mx[j] = a 
      mx_station[j] = row[3] + '(' + str(j+4) + ')' # 역 이름과 시간대
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.figure(dpi = 200)
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.xlabel('하차 시간이 가장 많은 역')
plt.ylabel('승객수')
plt.show()
```

## 모든 지하철 역에서 시간대별 승하차 인원을 모두 더하면

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/subwayTime.csv', 'r') # 인코딩이 utf-8이라면 생략 가능
data = csv.reader(f)
header_01 = next(data)
header_02 = next(data)

s_in = [0] *24
s_out = [0] *24

for row in data : 
  row[4:] = map(int, row[4:])
  for i in range(24) :
    s_in[i] += row[4 + i *2] # 승차
    s_out[i] += row[5 + i *2] # 하차

plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.figure(dpi = 200)

plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.xlabel('시간대')
plt.ylabel('승객수')
plt.show()
```
