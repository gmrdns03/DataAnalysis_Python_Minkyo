```python
import plotly.express as px
import pandas as pd
```

# 데이터 준비하기

```python
df = pd.read_csv('/content/도로교통공단_교통사고 정보 - 복사본.csv', encoding='euc-kr')
df.info()
```

```python
df.head()
```

```python
df.tail()
```

# 교통사고 시간 전처리

```python
# substr 문자열 자르기 위해 데이터 타입 변경하기
# 데이터프레임.astype({'컬럼': '데이터타입'})
df = df.astype({'발생년월일시': 'string'})  
```

```python
df.info()
```

```python
# 20190100100 -> 00 시간으로 분리
df['발생시간'] = df['발생년월일시'].str[8:]
```

```python
df.info()
```

```python
# 발생시간(str) -> 숫자형(int)으로 변경
df = df.astype({'발생시간' : 'int64'})
df.info()
```

```python
# 2019010100 -> 20190101 추출해서 -> 2019-01-01(날짜형태 YYYY-MM-DD 형태로 변경)
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8],
                                 format='%Y-%m-%d', errors='raise')  # 규칙 format='%Y-%m-%d'

df.info()
```

```python
df.head()
```

# 지도 활용하여 교통사고 현황 분석

```python
import folium
# location은 지도의 기준점을 지정해주는 것이다.
# 보통 내가 표시하고자 하는 지역의 시청을 기준으로 잡으면 된다.
map = folium.Map(location=[36.351, 127.386]) # 대전 시청좌표
map
```

## 지도에 필요한 정보만 담기 -> 대전의 교통사고 현황을 지도에 표출

```python
df_map = df[df.발생지시도 == '대전']
df_map.info()
```

### 버블(써클)을 그리기 위해서는 int64형이면 안된다. float64형으로 변환해줘야 한다.

## 사망자수, 부상자수, 중상자수, 경상자수 전처리

-> 지도에 원을 그리기 위해서

지도에 원을 그리는 것을 써클 마커라고 한다.

```python
df_map = df_map.astype({'사망자수' : 'float64'})
df_map = df_map.astype({'중상자수' : 'float64'})
df_map = df_map.astype({'부상자수' : 'float64'})
df_map = df_map.astype({'경상자수' : 'float64'})
df_map.info()
```

## 사망자수, 경상자 수에 따른 CircleMarker 지도에 표출하기

```python
for n in df_map.index :
  cnt = df_map['사망자수'][n] + df_map['경상자수'][n]  # 사망자수 + 경상자수 -> 써클 마커의 사이즈
  folium.CircleMarker([df_map['위도'][n], df_map['경도'][n]], radius= cnt * 10, popup=df_map['사고유형'][n], color='#3186cc', fill_color='#3186cc').add_to(map)

map
```

### 본인이 원하는 지역을 선택하여 my_map에 담아 데이터 분석하기

1. 교통사고 발생 시간 별/구 별 사망자 현황

2. 발생시간 별 사망자 현황

3. 사망자수, 경상자 수에 따른 circleMarker 지도에 표출하기

```python
import plotly.express as px
import pandas as pd
import folium

Goyang_map = folium.Map(location=[37.6583946277581, 126.83195674338178])

df = pd.read_csv('/content/도로교통공단_교통사고 정보 - 복사본.csv', encoding='euc-kr')
df = df.astype({'발생년월일시': 'string'})  
df['발생시간'] = df['발생년월일시'].str[8:]
df = df.astype({'발생시간' : 'int64'})
df['발생년월일시'] = pd.to_datetime(df['발생년월일시'].str[:8],
                                 format='%Y-%m-%d', errors='raise')  # 규칙 format='%Y-%m-%d'

df.info()
```

```python
Goyang_map
```

```python
my_map = df[df.발생지시군구 == '고양시']
my_map.info()
```

```python
my_map.head()
```

```python
# 경기도 교통사고 발생 시간 별/구 별 사망자 현황
gyungido_mqp = df[df.발생지시도 == '경기']
fig = px.bar(gyungido_mqp, x="발생시간", y= "사망자수", color='발생지시군구', title='경기도 교통사고 발생 시간별 사망자 현황')

fig.show()
```

```python
# 경기도 교통사고 발생 시간 별/구 별 사망자 현황
fig = px.bar(gyungido_mqp, x="사망자수", y= "발생시간", color='발생지시군구', orientation='h', title='경기도 시군구별 교통사고 사망자 현황')

fig.show()
```

```python
# 고양시 교통사고 발생 시간별 사망자 현황
fig = px.bar(my_map, x="발생시간", y= "사망자수", color='발생지시군구', title='고양시 교통사고 발생 시간별 사망자 현황')

fig.show()
```

```python
# 써클을 그리기 위한 전처리
my_map = my_map.astype({'사망자수' : 'float64'})
my_map = my_map.astype({'중상자수' : 'float64'})
my_map = my_map.astype({'부상자수' : 'float64'})
my_map = my_map.astype({'경상자수' : 'float64'})
my_map.info()
```

```python
my_map.head()
```

```python
for n in my_map.index :
  cnt = my_map['사망자수'][n] + my_map['경상자수'][n]  # 사망자수 + 경상자수 -> 써클 마커의 사이즈
  folium.CircleMarker([my_map['위도'][n], my_map['경도'][n]], radius= cnt * 10, popup=my_map['사고유형'][n], color='#3186cc', fill_color='#3186cc').add_to(Goyang_map)

Goyang_map
```
